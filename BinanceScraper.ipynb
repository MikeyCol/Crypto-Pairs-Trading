```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)

driver.get("https://www.binance.com/en/markets")
driver.implicitly_wait(5)
driver.find_element(by=By.CSS_SELECTOR, value = "button#onetrust-pc-btn-handler.cookie-setting-link").click()
driver.implicitly_wait(3)
driver.find_element(by=By.CLASS_NAME, value = "ot-pc-refuse-all-handler").click()
for i in range(8):
  soup = BeautifulSoup(driver.page_source, "lxml")
  tabContainer = soup.find(id="tabContainer")
  coinNamePrice = []
  names = tabContainer.find_all("div", class_="css-1qp5tsr")
  prices = tabContainer.find_all("div", class_="css-ydcgk2")
  for name,price in zip(names,prices):
    name = str(name)
    price = str(price)
    print(name)
    print(price)
    coinNamePrice.append((name[name.index('>')+1:name.index('</')],price[price.index('>')+1:price.index('</')]))
  print(coinNamePrice)
  nextPage = driver.find_elements(by=By.CSS_SELECTOR, value = "button.mirror.css-1w7smpm")
  if nextPage == None:
    break
  if len(nextPage) > 1:
    nextPage[1].click()
  else:
    nextPage[0].click()
with open('binanceMarketPrices.txt','w') as file:
	for coinInfo in coinNamePrice:
		file.write(str(coinInfo)+'\n')
```
